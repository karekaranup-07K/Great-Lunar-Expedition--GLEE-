## A] Thermal Properties of Lunar South Pole 
### 1. **Temperature Variations and Thermal Inertia**
- **Scenario:** Analyze the temperature data collected over a lunar day and night cycle.
- **Analysis:**
  - **Temperature Profiles:** Plot temperature variations over time to understand the diurnal temperature cycle.
  - **Thermal Inertia:** Calculate the rate of temperature change to infer the thermal properties of the lunar regolith.
  - **Heat Flow:** Determine the heat flow in the lunar surface layers by studying temperature gradients.

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

# Generate synthetic temperature data for a lunar day (14 Earth days) and night (14 Earth days)
# Assume 1-hour intervals, total 28 days = 28*24 hours
total_hours = 28 * 24
time = np.arange(total_hours)  # Time in hours

# Synthetic temperature profile
# Daytime temperatures (0-14 days) - range from -50 to 120 C
# Nighttime temperatures (14-28 days) - range from -170 to -50 C
temperature_day = np.linspace(-50, 120, 14 * 24)
temperature_night = np.linspace(-170, -50, 14 * 24)
temperature = np.concatenate((temperature_day, temperature_night))

# Add some random noise to simulate real data variations
noise = np.random.normal(0, 5, total_hours)
temperature += noise

# Create a DataFrame for better handling of data
data = pd.DataFrame({'Time (hours)': time, 'Temperature (C)': temperature})

# Plot temperature variations over time
plt.figure(figsize=(12, 6))
plt.plot(data['Time (hours)'], data['Temperature (C)'], label='Temperature (C)', color='b')
plt.xlabel('Time (hours)')
plt.ylabel('Temperature (C)')
plt.title('Temperature Variations over a Lunar Day and Night Cycle')
plt.legend()
plt.grid(True)
plt.show()

# Calculate the rate of temperature change (thermal inertia)
# Rate of change = dT/dt (where T is temperature and t is time)
data['Temperature Change Rate (C/hour)'] = data['Temperature (C)'].diff() / data['Time (hours)'].diff()

# Plot the rate of temperature change
plt.figure(figsize=(12, 6))
plt.plot(data['Time (hours)'], data['Temperature Change Rate (C/hour)'], label='Temperature Change Rate (C/hour)', color='r')
plt.xlabel('Time (hours)')
plt.ylabel('Temperature Change Rate (C/hour)')
plt.title('Rate of Temperature Change over a Lunar Day and Night Cycle')
plt.legend()
plt.grid(True)
plt.show()

# Calculate the heat flow in the lunar surface layers
# Assume a simple linear model: Heat flow (Q) = -k * (dT/dx)
# Where k is the thermal conductivity of the lunar regolith and dT/dx is the temperature gradient
# For simplicity, assume k = 1.0 W/mK and dx = 1.0 m
thermal_conductivity = 1.0  # W/mK
distance = 1.0  # m
data['Heat Flow (W/m^2)'] = -thermal_conductivity * data['Temperature Change Rate (C/hour)']

# Plot the heat flow
plt.figure(figsize=(12, 6))
plt.plot(data['Time (hours)'], data['Heat Flow (W/m^2)'], label='Heat Flow (W/m^2)', color='g')
plt.xlabel('Time (hours)')
plt.ylabel('Heat Flow (W/m^2)')
plt.title('Heat Flow in the Lunar Surface Layers')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Generate synthetic temperature data for two layers
surface_temperature = np.random.uniform(-50, 120, total_hours)
subsurface_temperature = surface_temperature - np.random.uniform(5, 15, total_hours)

# Calculate thermal conductivity using Fourier's law
# Q = -k * (dT/dx), where Q is heat flux, dT is temperature difference, dx is distance
distance = 0.1  # Assume 10 cm depth
heat_flux = -thermal_conductivity * (subsurface_temperature - surface_temperature) / distance

# Plot thermal conductivity estimation
plt.figure(figsize=(12, 6))
plt.plot(time, heat_flux, label='Estimated Thermal Conductivity (W/mK)')
plt.xlabel('Time (hours)')
plt.ylabel('Thermal Conductivity (W/mK)')
plt.title('Thermal Conductivity Estimation')
plt.legend()
plt.grid(True)
plt.show()
```

## B] **Magnetic Field Mapping**
- **Scenario:** Create a magnetic field map of the deployment area using the magnetometer data.
- **Analysis:**
  - **Magnetic Anomalies:** Identify and map magnetic anomalies, which could indicate variations in the lunar crust composition.
  - **Solar Wind Interaction:** Analyze how the magnetic field changes over time to study the interaction with the solar wind.
  - **Subsurface Structures:** Infer subsurface structures and composition by correlating magnetic field variations with surface features.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.ndimage import gaussian_filter

# Parameters for synthetic data
grid_size = 50  # Grid size for the deployment area
total_hours = 28 * 24  # Total hours in the dataset

# Generate synthetic magnetometer data (in microteslas)
np.random.seed(42)  # For reproducibility
magnetic_field_data = np.random.normal(0, 5, (grid_size, grid_size, total_hours))

# Simulate magnetic anomalies
anomaly_locations = [(20, 20), (35, 15), (10, 40)]
for loc in anomaly_locations:
    magnetic_field_data[loc[0]-2:loc[0]+3, loc[1]-2:loc[1]+3, :] += np.random.uniform(10, 30, (5, 5, total_hours))

# Apply Gaussian filter for smoothing (simulating measurement noise reduction)
smoothed_magnetic_field = gaussian_filter(magnetic_field_data, sigma=1)

# Magnetic Anomalies: Calculate mean magnetic field strength for each grid point
mean_magnetic_field = np.mean(smoothed_magnetic_field, axis=2)

# Plot magnetic field map with anomalies
plt.figure(figsize=(10, 8))
plt.imshow(mean_magnetic_field, cmap='coolwarm', extent=[0, grid_size, 0, grid_size])
plt.colorbar(label='Mean Magnetic Field (ŒºT)')
plt.title('Magnetic Field Map with Anomalies')
plt.scatter(*zip(*anomaly_locations), color='black', label='Anomaly Locations', marker='x')
plt.xlabel('Grid X Coordinate')
plt.ylabel('Grid Y Coordinate')
plt.legend()
plt.show()

# Temporal Analysis: Study how the magnetic field changes over time
selected_point = (25, 25)  # A point in the grid to analyze over time
time_series_data = smoothed_magnetic_field[selected_point[0], selected_point[1], :]

# Plot magnetic field changes over time
plt.figure(figsize=(12, 6))
plt.plot(np.arange(total_hours), time_series_data, label='Magnetic Field (ŒºT)')
plt.xlabel('Time (hours)')
plt.ylabel('Magnetic Field (ŒºT)')
plt.title('Magnetic Field Changes Over Time')
plt.legend()
plt.grid(True)
plt.show()

# Subsurface Structures: Infer subsurface structures (demonstration purposes)
# For a real scenario, this would involve complex modeling and correlation with surface data
subsurface_inference = mean_magnetic_field * np.random.uniform(0.8, 1.2, (grid_size, grid_size))

# Plot inferred subsurface structures
plt.figure(figsize=(10, 8))
plt.imshow(subsurface_inference, cmap='viridis', extent=[0, grid_size, 0, grid_size])
plt.colorbar(label='Inferred Subsurface Magnetic Field (ŒºT)')
plt.title('Inferred Subsurface Structures')
plt.xlabel('Grid X Coordinate')
plt.ylabel('Grid Y Coordinate')
plt.show()
```

## C] **Seismic Activity and Impact Detection**
- **Scenario:** Detect and analyze seismic events or impacts using accelerometer and gyroscope data.
- **Analysis:**
  - **Moonquake Detection:** Identify and categorize moonquakes by their magnitude and frequency.
  - **Impact Events:** Detect meteoroid impacts and analyze their effects on the lunar surface.
  - **Seismic Wave Propagation:** Study the propagation of seismic waves to understand the lunar interior structure.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import find_peaks, welch

# Parameters
total_hours = 28 * 24
sampling_rate = 100  # samples per second
total_samples = total_hours * 3600 * sampling_rate

# Generate synthetic accelerometer data (m/s^2) for seismic activity analysis
accelerometer_data = np.random.normal(0, 0.1, total_samples)

# Simulate moonquakes as sudden spikes in accelerometer data
moonquake_times = np.random.choice(total_samples, size=10, replace=False)
for time in moonquake_times:
    accelerometer_data[time:time+sampling_rate*5] += np.random.uniform(0.5, 2.0, sampling_rate*5)

# Simulate meteoroid impacts as distinct events
impact_times = np.random.choice(total_samples, size=5, replace=False)
for time in impact_times:
    accelerometer_data[time:time+sampling_rate*10] += np.random.uniform(1.0, 3.0, sampling_rate*10)

# Detect moonquakes and impacts using peak detection
peaks, _ = find_peaks(accelerometer_data, height=0.5)

# Plot accelerometer data with detected events
plt.figure(figsize=(12, 6))
plt.plot(accelerometer_data[:sampling_rate * 3600], label='Accelerometer Data')
plt.plot(peaks[:sampling_rate * 3600], accelerometer_data[peaks[:sampling_rate * 3600]], "x", label='Detected Events')
plt.xlabel('Time (seconds)')
plt.ylabel('Acceleration (m/s^2)')
plt.title('Seismic Activity Detection')
plt.legend()
plt.grid(True)
plt.show()

# Perform frequency analysis to study seismic wave propagation
frequencies, power_spectrum = welch(accelerometer_data, fs=sampling_rate, nperseg=1024)

# Plot power spectral density
plt.figure(figsize=(12, 6))
plt.semilogy(frequencies, power_spectrum)
plt.xlabel('Frequency (Hz)')
plt.ylabel('Power Spectral Density (PSD)')
plt.title('Seismic Wave Propagation Analysis')
plt.grid(True)
plt.show()
```
